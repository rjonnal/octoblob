```python
import numpy as np
from matplotlib import pyplot as plt

t = np.arange(-3,3,.001)
dtheta = np.pi/4
source = np.random.randn(len(t)*2) + 1j*np.random.randn(len(t)*2)

offset = 100
n1 = 0.1*np.random.randn(len(t))
n2 = 0.1*np.random.randn(len(t))

y1 = source[:len(t)]+n1
y2 = source[offset:len(t)+offset]+n2

plt.figure()
plt.plot(t,np.real(y1))
plt.plot(t,np.real(y2))

ccr = np.fft.ifft(np.fft.fft(y1)*np.conj(np.fft.fft(y2)))
acr = np.fft.ifft(np.fft.fft(np.abs(y1))*np.conj(np.fft.fft(np.abs(y2))))
plt.figure()
plt.plot(np.real(ccr))
plt.plot(np.real(acr))
plt.xlim((75,125))
plt.figure()
plt.plot(np.imag(ccr))
plt.plot(np.imag(acr))
plt.xlim((75,125))


plt.show()
```
